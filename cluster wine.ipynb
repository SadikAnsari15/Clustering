# wine-clustering

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import DBSCAN
```

# read the data

```python
data=pd.read_csv("wine-clustering.csv")
```

```python
data1=data.copy()
```

```python
data.info()
```

```python
data
```

```python
array=data.values
```

#  normalization the data

```python
from sklearn.preprocessing import StandardScaler
stscaler=StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
# eps= ramdon value,  min_sample=n+1  ,n=total number  of columns
dbscan = DBSCAN(eps=2.5, min_samples=14)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
clustered= pd.concat([data,cl],axis=1)
```

```python
clustered
```

```python
clustered[clustered['cluster']==-1]
#clustered[clustered['cluster']==0]
```

```python
# k  graph
from sklearn.neighbors import NearestNeighbors
from matplotlib import pyplot as plt


neighbors = NearestNeighbors (n_neighbors = 6) # minimum sample so we take 6 sample
neighbors_fit = neighbors.fit(X)
distances,indices = neighbors_fit.kneighbors(X)    # kneighbors  = 6


#plot thde graph


distances = np.sort(distances,axis = 0)
distances = distances[:,1]
plt.plot(distances)
```

```python
from sklearn.metrics import silhouette_score
%matplotlib inline
print("silhouette coifficent: %0.3f" %silhouette_score(X , dbscan.labels_))
```

```python
#this mean is with outleir
clustered.mean()
```

```python
# this is without outlier
clustered[clustered['cluster']==0].mean()
```

# K-Mean

```python
from sklearn.cluster import KMeans
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(data1)
df_norm.describe()
```

```python
from sklearn.cluster import KMeans
kmeans_clust = KMeans(n_clusters=3)
kmeans_clust.fit(df_norm)
```

```python
kmeans_clust.labels_
```

```python
df_norm.iloc[kmeans_clust.labels_==0,:]
#df_norm.iloc[kmeans_clust.labels_==1,:]
#df_norm.iloc[kmeans_clust.labels_==2,:]
```

```python
kmeans_clust.cluster_centers_[0].reshape(1,13).shape
```

```python
data1['clusterID']=pd.Series(kmeans_clust.labels_)

result = data1.iloc[:,0:].groupby(data1.clusterID).mean()
result
```

# find the K value

```python
WSS = [] # variable for storing total within sum of squares for each kmeans 
for i in range(2,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_norm)
    WSS.append(kmeans.inertia_) 
    print("TWSS when k is",i,sum(WSS))
plt.plot(range(2,11),WSS)
plt.title('the elbow mathod')
plt.xlabel('WSS')
plt.ylabel("no.clust")
plt.show()
```
